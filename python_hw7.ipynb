```python
# Filter out all of the empty strings from the list below

# Output: ['Argentina', 'San Diego', 'Boston', 'New York']

location = [" ", "Argentina", " ", "San Diego", " ", "Boston", " ", "New York", " "]

def countries(location):
    if location in location == " ":
        return False
    else:
        return True

print(list(filter(countries,location)))
```

```python
# Exercise #2
# Write an anonymous function that sorts this list by the last name...
# Hint: Use the ".sort()" method and access the key"

# Output: ['Victor aNisimov', 'Gary A.J. Bernstein', 'Joel Carter', 'Andrew P. Garfield', 'David hassELHOFF']

actors = ['David hassELHOFF', 'Andrew P. Garfield', 'Gary A.J. Bernstein', 'Victor aNisimov', 'Joel Carter']

def lastname(actors):
    actors.sort(key=lambda x: x.split(' ')[-1].lower())
    return actors

print(lastname(actors))
```

```python
# Exercise #4
# Write a recursion function to perform the fibonacci sequence up to the number passed in.

# Output for fib(5) => 
# Iteration 0: 1
# Iteration 1: 1
# Iteration 2: 2
# Iteration 3: 3
# Iteration 4: 5
# Iteration 5: 8

def fibo(n):
    if n == 0:
        return 0
    elif n == 0:
        return 0
    elif n == 1 or n == 2:
        return 1
    else:
        return fibo(n-1) + fibo(n-2)

def fiboN(n):
    for i in range(0, n+1):
        print(f"Interation {i}: {fibo(i)}")
        
print(fiboN(5))
```
